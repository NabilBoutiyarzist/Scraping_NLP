Lecture du fichier texte :

```python
chemin = "Articles_Occidentaux.txt"
f = open(chemin, "r",encoding="utf-8")
textebrut = f.read()
textebrut = re.sub(r"[.:' ;,?=!\n]+"," ",textebrut) 
```

Traitement du texte brut :

```python
import treetaggerwrapper
tagger=treetaggerwrapper.TreeTagger(TAGDIR="C:/TreeTagger/" ,TAGLANG='fr')

u=textebrut.split(' ')
print (u)

tags=tagger.tag_text(u)
print (tags)
```

Export vers excel des Graphies Pos Lemme

```python
import xlsxwriter
LH = xlsxwriter.Workbook('C:/Users/10172919/Desktop/M2/Stats textuelles/NLP/TreeTagger_Occident.xlsx')
export= LH.add_worksheet()

row=1
col=0
export.write(0,0,'Graphies')
export.write(0,1,'POS')
export.write(0,2,'Lemmes')

for element in tags:
    try :
        extrait_un=element.split('\t')[0]
        extrait_deux=element.split('\t')[1]
        extrait_trois=element.split('\t')[2]
        export.write(row,col,extrait_un)
        export.write(row,col+1,extrait_deux)
        export.write(row,col+2,extrait_trois)
        row += 1
        col=0  
        
    except Exception : 
        continue

LH.close()
```

Comptage des mots les plus frÃ©quent avec les Lemmes

```python
motot = {}
liste_lemme = []
for i in tags:
    liste_lemme.append(i.split('\t')[2])

for mot in liste_lemme:
    motot[mot] = 0
for mot in liste_lemme:
    motot[mot] += 1

import csv

# Nom du fichier CSV de sortie
nom_fichier_csv = 'mot_plus_frequent_occident.csv'

# Ã‰criture du dictionnaire dans le fichier CSV
with open(nom_fichier_csv, 'w', newline='', encoding='utf-8-sig') as fichier_csv:
    # CrÃ©ation de l'objet writer
    writer = csv.writer(fichier_csv)

    # Ã‰criture des en-tÃªtes
    writer.writerow(['ClÃ©', 'Valeur'])

    # Ã‰criture des donnÃ©es du dictionnaire dans le fichier CSV
    for cle, valeur in motot.items():
        writer.writerow([cle, valeur])
    
```
